# Exercice 1 : Analyse Multi-Variables (FacetGrid)
**Objectif :** Comparer des distributions sur plusieurs catégories simultanément.

**Scénario :** Tu travailles sur un dataset de performances de vente.

1. Génère ou utilise un DataFrame avec les colonnes : Ventes, Produit (A, B, C) et Region (Nord, Sud).

2. Utilise sns.FacetGrid pour créer une grille de graphiques où chaque colonne de la grille représente une Region.

3. À l'intérieur de cette grille, affiche un Histogramme (sns.histplot) des Ventes avec une couleur différente par Produit.

4. Ajoute une légende globale et un titre à la figure.



```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# --- Étape 1 : Génération du DataFrame ---
np.random.seed(42) # Pour obtenir les mêmes résultats aléatoires à chaque exécution
n_samples = 500

df = pd.DataFrame({
    'Ventes': np.random.normal(loc=150, scale=30, size=n_samples), # Distribution normale
    'Produit': np.random.choice(['A', 'B', 'C'], size=n_samples),  # Choix aléatoire
    'Region': np.random.choice(['Nord', 'Sud'], size=n_samples)    # Choix aléatoire
})

# --- Étape 2 : Création de l'Analyse Multi-Variables avec FacetGrid ---
sns.set_theme(style="whitegrid") # Ajoute un fond quadrillé plus esthétique

# 2.1 Initialisation de la grille
g = sns.FacetGrid(df, col="Region", hue="Produit", height=5, aspect=1.2, palette="muted")

# 2.2 Remplissage de la grille avec les histogrammes
g.map(sns.histplot, "Ventes", alpha=0.6, bins=20)

# 2.3 Ajout de la légende globale
g.add_legend(title="Catégories de Produits")

# 2.4 Ajout du titre global
g.fig.suptitle("Distribution des Ventes par Région et par Produit", y=1.05, fontsize=16)

# Affichage du graphique (ou plt.savefig('nom.png') pour sauvegarder)
plt.show()
```
