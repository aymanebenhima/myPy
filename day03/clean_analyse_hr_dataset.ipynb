# Exercice 1 : Nettoyage et Analyse de Données RH

1. importer pandas

```python
import pandas as pd
```

2. Créer le DataFrame 

```python
data = [
    {"nom": "Alice", "departement": "IT", "salaire": 4500, "date_embauche": "2020-01-15"},
    {"nom": "Bob", "departement": "RH", "salaire": None, "date_embauche": "2021-03-22"},
    {"nom": "Charlie", "departement": "IT", "salaire": 5200, "date_embauche": "2019-11-05"},
    {"nom": "David", "departement": "Finance", "salaire": 3800, "date_embauche": None}
]

df = pd.DataFrame(data)

print(df)
```

3. Remplir les salaires manquants par la moyenne de l'entreprise

```python
moy_salary = df['salaire'].mean()
df["salaire"] = df["salaire"].fillna(moy_salary)
print(df)
```

**Explication:**
- mean() calcule la moyenne des salaires existants
- fillna() remplace les valeurs None / NaN

4. Supprimer les lignes sans date d'embauche

```python
df = df.dropna(subset=["date_embauche"])
print(df)
```

**Explication:**
- subset=["data_embauche"] indique la colonne à vérifier
- Toute ligne avec date manquante est supprimée

5. Ajouter la colonne prime (10%)

```python
df["prime"] = df["salaire"] * 0.1
print(df)
```

6. Filter les employés IT avec salaire > 5000

```python
res = df[
    (df["departement"] == "IT") &
    (df["salaire"] > 5000)
]

print(res)
```
