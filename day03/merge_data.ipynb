# Exercice: Fusion de données (Merge & Join)

1. Importer pandas

````python
import pandas as pd
````

2. Créer les DataFrames

````python
df_clients = pd.DataFrame([
    {"ID_Client": 1, "Nom": "Alice", "Email": "alice@email.com"},
    {"ID_Client": 2, "Nom": "Bob", "Email": "bob@email.com"},
    {"ID_Client": 3, "Nom": "Charlie", "Email": "charlie@email.com"},
    {"ID_Client": 4, "Nom": "David", "Email": "david@email.com"}
])

df_commandes = pd.DataFrame([
    {"ID_Commande": 101, "ID_Client": 1, "Montant": 250},
    {"ID_Commande": 102, "ID_Client": 1, "Montant": 400},
    {"ID_Commande": 103, "ID_Client": 3, "Montant": 150}
])
````

3. Inner Join: Clients avec leurs commandes

````python
commandes_clients = pd.merge(
    df_clients,
    df_commandes,
    on="ID_Client",
    how="inner"
)

print(commandes_clients)
````

4. Identifier les clients sans commande (Left Join)

````python
clients_avec_commandes = pd.merge(
    df_clients,
    df_commandes,
    on="ID_Client",
    how="left"
)

clients_sans_commandes = clients_avec_commandes[clients_avec_commandes["ID_Commande"].isna()]
print(clients_sans_commandes)
````

## Explication des Types de Join

### Inner Join
- Garde uniquement les lignes présentes dans les deux DataFrames

```python
how="inner"
```

### Left Join
- Garde tous les clients, même ceux sans commande

```python
how="left"
```
