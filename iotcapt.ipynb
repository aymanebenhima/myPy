# Exercice 1 : Analyse de Données Capteur (Variables & Types)
- **Contexte :** Vous recevez des données brutes d'un capteur IoT sous forme de chaînes de caractères. Vous devez les nettoyer et les convertir.

**Instructions :**

1. Créez trois variables reçues d'une API : raw_temp = " 22.5 " (avec espaces), raw_humidity = "45" et sensor_status = "Active".

```python
# Raw data from API
raw_temp = " 22.5 "
raw_humidity = "45"
sensor_status = "Active"
```

2. Nettoyez raw_temp en supprimant les espaces et convertissez-la en float.
3. Convertissez raw_humidity en int.

```python
# 2. Clean and convert
temperature = float(raw_temp.strip())
humidity = int(raw_humidity.strip())

print(f"Temperature: {temperature} °C")
print(f"Humidity: {humidity} %")
```

4. Créez une variable booléenne is_alert qui vaut True si la température est supérieure à 25 ou si l'humidité est supérieure à 50.

```python
# 3. Alert condition
is_alert = sensor_status == "Active" and (temperature > 30 or humidity > 70)

print(f"Alert: {'YES' if is_alert else 'NO'}")
```

5. Affichez un rapport formaté : "Capteur [STATUS] - Temp: [X]°C | Alerte: [YES/NO]".

```python
# 4. Formatted report
alert_label = "YES" if is_alert else "NO"
print(f"Capteur {sensor_status} - Temp: {temperature} °C, Humidité: {humidity} % - Alert: {alert_label}")
```

6. Essayez de modifier une chaîne de caractères directement (ex: sensor_status[0] = "I"). Constatez que les strings sont immuables en Python.

```python
try:
    sensor_status[0] = "I"
except TypeError as e:
    print(f"Error: {e} - Sensor status is immutable.")
```
