# Exercice : Gestion d'Inventaire (Listes & Boucles For/While)
**Contexte :** Vous gérez un stock de composants pour un projet robotique.

**Instructions :**
1. Créez une liste stock contenant : "Moteur", "Capteur", "Batterie", "Câble", "LED".

```python
# 1. Initial stock list
stock = ["Moteur", "Capteur", "Batterie", "Cable", "LED"]
```

2. Ajoutez "Micro-contrôleur" à la fin et "Châssis" au début de la liste.

```python
# 2. Add new item
stock.append("Micro-contrôleur")
stock.insert(0, "Châssis")

stock
```

3. Utilisez une boucle while pour demander à l'utilisateur de saisir un composant à rechercher. La boucle s'arrête si l'utilisateur tape "stop".

```python
# 3 & 4. While Loop - search for a component
while True:
    composant = input("Search for a component (or 'stop' to quit): ")
    if composant == "stop":
        break
    if composant in stock:
        print(f"{composant} is in the stock with quantity {stock.count(composant)} and index {stock.index(composant)}.")
    else:
        print(f"{composant} is not in the stock.")
```

5. Utilisez une List Comprehension pour créer une nouvelle liste stock_court contenant uniquement les composants dont le nom fait moins de 6 caractères.

```python
# 5. List comprehension names shorter than 6 characters
stock_count = [item for item in stock if len(item) < 6]
print(f"Components with names shorter than 6 characters: {stock_count}")
```

6. Affichez la liste finale triée par ordre alphabétique.

```python
# 6. Sorted alphabetically
print(f"Stock sorted alphabetically: {sorted(stock)}")

## sorting in reverse order
print(f"Stock sorted in reverse order: {sorted(stock, reverse=True)}")
```

# Exercice 1 : Annuaire de Serveurs (Dictionnaires & Nesting)
**Contexte :** Vous devez structurer les informations d'une infrastructure de serveurs.

**Instructions :**

1. Créez un dictionnaire infrastructure où chaque clé est un nom de serveur (srv_01, srv_02) et chaque valeur est un autre dictionnaire contenant : ip, role (Web, DB, ou Cache), et status (True/False).

```python
# 1. Create a dictionary
infrastructure = {
    "srv_01": {
        "ip": "192.168.1.10",
        "role": "Web",
        "status": True
    },
    "srv_02": {
        "ip": "192.168.1.11",
        "role": "DB",
        "status": False
    }
}
```

2. Ajoutez dynamiquement un troisième serveur srv_03 à l'infrastructure.

```python
# 2. Add dynamically new server

def add_server(infra, name, ip, role, status=True):
    infra[name] = {
        "ip": ip,
        "role": role,
        "status": status
    }

add_server(infrastructure, "srv_03", "192.168.1.12", "Cache")
add_server(infrastructure, "srv_04", "192.168.1.15", "Web", status=False)
print(infrastructure)
```

3. Écrivez une boucle qui parcourt l'infrastructure et affiche uniquement les adresses IP des serveurs dont le rôle est "Web".

```python
# 3. Display IPs of Web servers
def display_web_servers(infra):
    print("Web Servers:")
    for srv_id, details in infra.items():
        if details["role"] == "Web":
            print(f"{srv_id}: {details['ip']}")
def get_ips_by_role(infra, role):
    return [details["ip"] for details in infra.values() if details["role"] == role]

web_ips = get_ips_by_role(infrastructure, "Web")
print(f"Web servers IPs: {web_ips}")

##
print(f"Web Servers: {display_web_servers(infrastructure)}")
```

4. Modifiez le statut de tous les serveurs pour les passer à False (maintenance) en une seule boucle.

```python
def set_maintenance(infra):
    for info in infra.values():
        info["status"] = False

set_maintenance(infrastructure)
print(infrastructure)
```

5. Extraction : Créez une liste contenant tous les rôles uniques présents dans l'infrastructure (utilisez un set pour éliminer les doublons).

```python
def get_unique_roles(infra):
    return list({info["role"] for info in infra.values()})

unique_roles = get_unique_roles(infrastructure)
print(f"Unique roles in infrastructure: {unique_roles}")

print(infrastructure)
```

## Exercice 2: Pipeline de Transformation (Tuples & Fonctions)
**Contexte :** Simulation d'un pipeline de données qui transforme des coordonnées GPS.

**Instructions :**

1. Créez une liste de tuples nommée points_gps : [(33.57, -7.58), (48.85, 2.35), (34.02, -6.83)].

```python
# 1. GPS points list
points_gps = [(33.57, -7.58), (48.85, 2.35), (34.02, -6.83)]
print(points_gps)
```

2. Créez une fonction format_coord(point) qui prend un tuple en entrée, utilise l'unpacking pour extraire lat et lon, et retourne une chaîne : "Lat: 33.5, Lon: -7.5" (arrondi à 1 décimale).

```python
# 2. Format a single coordinate
def format_coord(point):
    lat, lon = point
    return f"Lat: {round(lat, 1)}, Long: {round(lon, 1)}"

print([format_coord(point) for point in points_gps])
```

3. Parcourez la liste points_gps et stockez les chaînes formatées dans une nouvelle liste.

```python
# 3. Format all points
def format_all(points):
    return [format_coord(p) for p in points]

print(format_all(points_gps))
```

**Défi Immuabilité :** Tentez de changer la latitude du premier point directement dans la liste. Proposez une solution pour "modifier" un tuple (indice : il faut recréer un nouveau tuple).

```python
# 4. Demonstrate tuple immutability
def demo_immutability(points):
    try:
        points[0][0] = 99.99
    except TypeError as e:
        print(f"Error: {e} - Points are immutable.")

demo_immutability(points_gps)
```

```python
# 5. "Modify" a tuple by recreating it
def modify_tuple(points, index, new_lat=None, new_lon=None):
    lat, lon = points[index]
    updated = (new_lat if new_lat is not None else lat,
              new_lon if new_lon is not None else lon)
    return points[:index] + [updated] + points[index+1:]

print(modify_tuple(points_gps, 1, new_lat=40.71))
```
